# Variáveis

```python
# Exemplo básico de declaração de variável

idade = 29
altura = 1.80
nome_sobrenome = "Rafaela Tiossi"

print(idade)
print(altura)
print(nome_sobrenome)
```

```python
# Esse dado é do tipo inteiro:
print(type(idade))
```

```python
# Esse dado é do tipo fracionário:
print(type(29))
```

```python
# Esse dado é do tipo conjunto de caracteres/textos:
print(type(nome_sobrenome))
```
