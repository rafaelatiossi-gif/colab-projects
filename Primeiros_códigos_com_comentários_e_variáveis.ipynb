```python
# Calculando a soma de dois números.

# Estrutura sequencial
 # Primeiro passo?
print ("Digite dois números:")

# Entrada
num1 = int(input("Número 1:"))
num2 = int(input("Número 2:"))

# Processamento
soma = num1 + num2

# Saída
print("A soma é: ", soma)
```

```python
# Verificar se um número é par ou ímpar
# Entrada - Nesse caso a variável é um número:
numero = int(input("Digite um número:"))

# Processamento e Saída

if numero % 2 == 0:
  print("O número é par.")
else:
  print("O número é ímpar.")
```

# Lambda e built-in functions

```python
dobro = lambda x: x * 2
print(dobro)
```

```python
# String
primeira_letra = lambda s: s[0]
print(primeira_letra("Rafaela"))
```

```python
# String
lista = [1,2,3,4,5]
print(lista[0])
```

```python
# Lista - Sorteio
lista = [(1, 3), (4, 1), (2, 3)]
lista.sort(key=lambda x: x[1])
print(lista)
```
