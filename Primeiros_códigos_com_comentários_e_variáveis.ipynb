# Organizando Código com Funções e Modularidade

```python
# Modularização é basicamente dividir o nosso programa em partes ou em pequenas funções

# APP ESTRUTURADO PARA CALCULAR A MÉDIA DOS ALUNOS

# 1. Passo: Estruturar o programa:
  # 1. Passo: Capturar notas dos alunos
  # 2. Passo: Calcular a média
  # 3. Passo: Exibir o resultado

  # 2. Passo: Definir as funções

  # - capturar_dados()

  # - calcular_media()

  # - exibir_resultado ()
```

```python
# Função para capturar dados
def capturar_dados():
  notas = []
  for i in range(3):
    nota = float(input(f"Digite a nota do aluno {i+1}: "))
    notas.append(nota)
  return notas

# Função calcular média:
def calcular_media(notas):
    soma = sum(notas)
    media = soma / len(notas)
    return media

# Função para exibir o resultado:
def exibir_resultado(media):
  if media >= 7:
    print(f"A média do aluno é {media}. Aprovado!")
  else:
    print(f"A média do aluno é {media}. Reprovado!")
```

```python
notas = capturar_dados()
media = calcular_media(notas)
exibir_resultado(media)
```

# Manipulação de arquivos

```python
# TXT é uma extensão para arquivos de texto
with open("arquivo.txt", "w") as arquivo:
  # Escrever alguns dados:
  arquivo.write("Linha 1 : Este é um exemplo de texto. \n")
  arquivo.write("Linha 2 : Este é um exemplo de texto. \n")
  arquivo.write("Linha 3 : Este é um exemplo de texto. \n")

  print("Dados gravados com sucesso!")
```

```python
# Manipulando arquivos txt
with open("arquivo.txt", "r") as arquivo:
  # Ler o conteúdo do arquivo:
  conteudo = arquivo.read()
  print(conteudo)
```

```python
# Apagar uma linha
with open("arquivo.txt", "r") as arquivo:
  linhas = arquivo.readlines()
  linhas.pop(1)
  print (linhas)
```
